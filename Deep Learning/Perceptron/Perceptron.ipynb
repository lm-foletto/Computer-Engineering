```python
import numpy as np

def sigmoid(x):
  return 1.0 / (1 + np.exp(-x))

'''class Perceptron():
  def __init__(self, n_inputs=4, n_hidden=8, n_outputs=3):
    # Sinapses
    self.w_ih = np.random.random((n_hidden, n_inputs))
    self.w_ho = np.random.random((n_outputs, n_hidden))
    # Biases
    self.b_hid = np.random.random((n_hidden, 1))
    self.b_out =  np.random.random((n_outputs, 1))

  def compute(self, inputs):
    # Garante que a entrada seja um vetor coluna
    inputs = np.reshape(inputs,(len(inputs),1))
    # Calcula as ativações, camada por camada
    self.s_hid = np.dot(self.w_ih, inputs) + self.b_hid
    self.z_hid = sigmoid(self.s_hid)
    self.s_out = np.dot(self.w_ho, self.z_hid) + self.b_out
    self.z_out = sigmoid(self.s_out)
    return self.z_out'''

class Perceptron():
    def __init__(self, n_inputs=3, n_hidden=4, n_outputs=3, ih_weights=None, ho_weights=None, b_hid=None, b_out=None):
        if ih_weights is None:
            self.w_ih = np.random.random((n_hidden, n_inputs))
        else:
            self.w_ih = ih_weights

        if ho_weights is None:
            self.w_ho = np.random.random((n_outputs, n_hidden))
        else:
            self.w_ho = ho_weights

        if b_hid is None:
            self.b_hid = np.random.random((n_hidden, 1))
        else:
            self.b_hid = b_hid

        if b_out is None:
            self.b_out = np.random.random((n_outputs, 1))
        else:
            self.b_out = b_out

    def compute(self, inputs):
        inputs = np.reshape(inputs, (len(inputs), 1))
        self.s_hid = np.dot(self.w_ih, inputs) + self.b_hid
        self.z_hid = sigmoid(self.s_hid)
        self.s_out = np.dot(self.w_ho, self.z_hid) + self.b_out
        self.z_out = sigmoid(self.s_out)
        return self.z_out

# Definindo os parâmetros especificados
ih = np.array([[1.12, 0.92, 1.28],
              [-0.88, -1.02, -1.46],
              [1.06, 0.63, 0.38],
              [-1.5, -1.99, -2.31]])

ho = np.array([[0.54, 2.51, -1.81, 5.15],
              [-6.22, -4.08, 1.75, -3.47],
              [5.89, 1.58, -1.72, -4.06]])

b_hid = np.array([[-8.86], [4.36], [-1.87], [4.79]])
b_out = np.array([[-0.82], [0.76], [-1.35]])

# Criando a instância da rede neural com os parâmetros específicos
p = Perceptron(n_inputs=3, n_hidden=4, n_outputs=3, ih_weights=ih, ho_weights=ho, b_hid=b_hid, b_out=b_out)
#p = Perceptron()

# Calculando as saídas para os conjuntos de entrada
output_1 = p.compute([1, 1, 1])
output_2 = p.compute([2, 2, 2])
output_3 = p.compute([3, 3, 3])
#p.compute([1,2,3,4])

# Arredondando as saídas para uma casa decimal
output_1 = np.round(output_1, 1)
output_2 = np.round(output_2, 1)
output_3 = np.round(output_3, 1)

# Saídas para os três conjuntos de entrada
print(output_1)
print(output_2)
print(output_3)


import matplotlib.pyplot as plt
import networkx as nx

# Criar um gráfico direcionado
G = nx.DiGraph()

# Adicionar os nós da rede (neurônios) em ordem: Entradas, Camadas Ocultas e Saídas
nodes = ["Entrada 1", "Entrada 2", "Entrada 3", "Entrada 4", "Camada 1 Neurônio 1", "Camada 1 Neurônio 2", "Camada 1 Neurônio 3",
         "Camada 1 Neurônio 4", "Camada 1 Neurônio 5", "Camada 2 Neurônio 1", "Camada 2 Neurônio 2", "Camada 2 Neurônio 3",
         "Camada 2 Neurônio 4", "Camada 2 Neurônio 5", "Saída 1", "Saída 2", "Saída 3"]
G.add_nodes_from(nodes)

# Adicionar as arestas com pesos (conexões e pesos)
edges = [("Entrada 1", "Camada 1 Neurônio 1", 1), ("Entrada 2", "Camada 1 Neurônio 1", 2), ("Entrada 3", "Camada 1 Neurônio 1", 3),
         ("Entrada 4", "Camada 1 Neurônio 1", 4), ("Entrada 1", "Camada 1 Neurônio 2", 5), ("Entrada 2", "Camada 1 Neurônio 2", 6),
         ("Entrada 3", "Camada 1 Neurônio 2", 7), ("Entrada 4", "Camada 1 Neurônio 2", 8), ("Entrada 1", "Camada 1 Neurônio 3", 9),
         ("Entrada 2", "Camada 1 Neurônio 3", 10), ("Entrada 3", "Camada 1 Neurônio 3", 11), ("Entrada 4", "Camada 1 Neurônio 3", 12),
         ("Entrada 1", "Camada 2 Neurônio 1", 13), ("Entrada 2", "Camada 2 Neurônio 1", 14), ("Entrada 3", "Camada 2 Neurônio 1", 15),
         ("Entrada 4", "Camada 2 Neurônio 1", 16), ("Camada 1 Neurônio 1", "Saída 1", 17), ("Camada 1 Neurônio 2", "Saída 1", 18),
         ("Camada 1 Neurônio 3", "Saída 1", 19), ("Camada 1 Neurônio 4", "Saída 2", 20), ("Camada 1 Neurônio 5", "Saída 2", 21),
         ("Camada 2 Neurônio 1", "Saída 2", 22), ("Camada 2 Neurônio 2", "Saída 3", 23), ("Camada 2 Neurônio 3", "Saída 3", 24),
         ("Camada 2 Neurônio 4", "Saída 1", 25), ("Camada 2 Neurônio 5", "Saída 2", 26)]

G.add_weighted_edges_from(edges)

# Obter posições dos nós para plotagem
pos = nx.spring_layout(G, seed=42)

# Obter pesos das arestas (conexões) para plotagem
edge_labels = {(u, v): d for u, v, d in G.edges(data="weight")}

# Plotar a rede com rótulos de pesos
plt.figure(figsize=(12, 8))
nx.draw(G, pos, with_labels=True, node_size=800, node_color="skyblue", font_size=8, font_weight="bold")
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=8, font_color="red")
plt.title("Rede Neural - Perceptron")
plt.show()
```
