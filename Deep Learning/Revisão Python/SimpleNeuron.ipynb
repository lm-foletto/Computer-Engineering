```python
import numpy as np
from matplotlib import pyplot as plt

# Função sigmoid para normalização
# Sigmoid function for normalization
def sigmoid(x):
    return 1.0 / (1 + np.exp(-x))

# Geração de valores de x de -5 a 5
# Generating x values from -5 to 5
x = np.linspace(-5.0, 5.0)

# Aplicação da função sigmoid aos valores de x
# Applying the sigmoid function to x values
y = sigmoid(x)

# Plotando a função sigmoid
# Plotting the sigmoid function
plt.plot(x, y)

# Classe Neuron com função de ativação sigmoid
# Neuron class with sigmoid activation function
class Neuron:
    def __init__(self, input_size=5):
        self.w = np.random.random(input_size)  # Inicialização dos pesos
        self.b = np.random.random()  # Inicialização do bias
    def compute(self, inputs):
        s = np.dot(self.w, inputs) + self.b  # Soma ponderada dos inputs
        z = sigmoid(s)  # Aplicação da função sigmoid
        return z

# Classe NeuronTanh com função de ativação tanh
# NeuronTanh class with tanh activation function
class NeuronTanh:
    def __init__(self, input_size=3):
        self.w = np.array([-1, 0, 1])  # Inicialização dos pesos
        self.b = -0.5  # Inicialização do bias

    def compute(self, inputs):
        s = np.dot(self.w, inputs) + self.b  # Soma ponderada dos inputs
        z = np.tanh(s)  # Aplicação da função tanh
        return z

# Criar uma instância do neurônio NeuronTanh
# Create an instance of NeuronTanh
n = NeuronTanh()
# n = Neuron()

# Calcular a saída para as entradas 1, 20, 3
# Calculate the output for inputs 1, 20, 3
n.compute([1, 20, 3])
# n.compute([1, 2, 3, 4, 5])
```
