O objetivo de usar a função de custo em deep learning é avaliar quão bem o modelo de rede neural está performando em uma tarefa específica, como classificação, regressão ou outra tarefa de aprendizado supervisionado. A função de custo, também conhecida como função erro ou recomensa, descreve a discrepância entre as previsões do modelo e os valores reais dos dados de treinamento. A importância da função de custo reside nos seguintes aspectos:

1. **Otimização do Modelo:** Durante o treinamento de uma rede neural, o objetivo é ajustar os pesos e os viéses da rede de forma a minimizar a função de custo. Isso significa que o modelo está tentando encontrar os parâmetros que resultam nas previsões mais precisas possíveis em relação aos dados de treinamento.

2. **Feedback para o Modelo:** A função de custo fornece feedback ao modelo. Quando a função de custo é alta, isso indica que o modelo está fazendo previsões incorretas e precisa ajustar seus parâmetros para melhorar o desempenho.

3. **Regularização:** Em muitos casos, a função de custo pode conter termos de regularização que ajudam a evitar o overfitting (ajustamento excessivo) do modelo aos dados de treinamento. Esses termos incentivam a simplicidade do modelo, ajudando a evitar a complexidade excessiva que pode levar a um desempenho pior em dados não vistos.

4. **Escolha da Arquitetura da Rede:** A escolha da função de custo está relacionada à tarefa que a rede neural está realizando. Diferentes tarefas podem exigir funções de custo específicas. Por exemplo, para classificação binária, pode-se usar a entropia cruzada binária, enquanto a regressão pode envolver o erro quadrático médio.

5. **Avaliação do Desempenho:** Após o treinamento, a função de custo também é usada para avaliar o desempenho do modelo em dados de validação ou teste. Modelos com valores mais baixos de função de custo tendem a fazer previsões mais precisas.

Em resumo, a função de custo desempenha um papel fundamental no treinamento e avaliação de modelos de deep learning. Seu objetivo é guiar o processo de otimização, ajudando a encontrar os melhores parâmetros do modelo para a tarefa em questão e, em última instância, melhorar o desempenho do modelo em dados reais.

```python
#Função cujo entrada deve possuir n variáveis para uma única saída
def func(x, y):
  return (x**2) * np.sin(y)
```

Usando o gradiente, é possível realizar várias tarefas, como otimização. Por exemplo, em algoritmos de otimização, como o Gradiente Descendente, o gradiente é usado para ajustar os parâmetros de um modelo de forma a minimizar (ou maximizar) uma função de custo. A direção do gradiente indica em qual direção a função de custo está diminuindo mais rapidamente, tornando-o valioso para encontrar os mínimos ou máximos da função.

Em resumo, o gradiente é uma ferramenta importante em matemática e em algoritmos de otimização, auxiliando na compreensão da taxa de variação das funções em relação às variáveis independentes e facilitando a otimização de modelos e processos.

```python
#Criando função gradiente
def grad(x,y):
  return np.array([2*x*sin(y), (x**2)*np.cos(y)])
```

```python
# np.linspace(-5.0, 5.0, 100) gera um array com 100 valores igualmente espaçados no intervalo de -5.0 a 5.0. Isso cria uma sequência de números que representa valores de x.
# np.ones(100) cria um array de 100 uns. Isso é usado para criar uma matriz de uns, que será usada para construir a matriz x e a matriz y.
# np.outer(np.linspace(-5.0, 5.0, 100), np.ones(100)) utiliza a função np.outer para criar uma matriz bidimensional x. A matriz x terá os valores de np.linspace(-5.0, 5.0, 100) nas linhas e os valores de np.ones(100) nas colunas.
x = np.outer(np.linspace(-5.0, 5.0, 100), np.ones(100))
y = np.outer(np.ones(100), np.linspace(-5.0, 5.0, 100))
z = func(x,y)
```

```python
# ax = plt.axes(projection='3d'): Cria um objeto de eixos 3D para a plotagem tridimensional.
ax = plt.axes(projection='3d')
# ax.view_init(50, 30): Define a vista da figura tridimensional com ângulos de inclinação de 50 graus e azimute de 30 graus.
ax.view_init(50, 30)
# ax.plot_surface(x, y, z, cmap=cm.coolwarm): Plota a superfície 3D com base nas matrizes x, y, e z, usando a paleta de cores "coolwarm".
ax.plot_surface(x, y, z, cmap=cm.coolwarm)
# Define o rótulo do eixo x como "Eixo x".
ax.set_xlabel('Eixo x')
# Define o rótulo do eixo y como "Eixo y".
ax.set_ylabel('Eixo y')
# Define o rótulo do eixo x como "Eixo x".
ax.set_zlabel('Eixo z')
```

```python
ax = plt.axes()
ax.imshow(-z.T, cmap=cm.coolwarm, extent=(-5, 5, -5, 5), interpolation='bilinear')
# Define o rótulo do eixo x como "Eixo x".
ax.set_xlabel('Eixo x')
# Define o rótulo do eixo y como "Eixo y".
ax.set_ylabel('Eixo y')
```

```python
from math import sin

qx = []
qy = []
qu = []
qv = []

for xi in np.linspace(-1.5*np.pi, 1.5*np.pi, 15):
  for yi in np.linspace(-1.5*np.pi, 1.5*np.pi, 15):
    ui, vi = grad(xi,yi)
    qx.append(xi)
    qy.append(yi)
    qu.append(ui)
    qv.append(vi)
```

```python
ax = plt.axes()
ax.imshow(-z.T, cmap=cm.coolwarm, extent=(-5, 5, -5, 5), interpolation='bilinear')
Q = ax.quiver(qx, qy, qu, qv, scale=360.0, pivot='mid')
qk = ax.quiverkey(Q, 0.1, 0.1, 0.1, '', labelpos='E', coordinates='figure')
# Define o rótulo do eixo x como "Eixo x".
ax.set_xlabel('Eixo x')
# Define o rótulo do eixo y como "Eixo y".
ax.set_ylabel('Eixo y')
```

```python
def func3d(x,y):
  f = 0.8
  return np.cos(x*f) * (25 - x**2) - np.cos(y*f) * (25 - y**2)
```

```python
x = np.outer(np.linspace(-5.0, 5.0, 100), np.ones(100))
y = np.outer(np.ones(100), np.linspace(-3.0, 7.0, 100))
z = func3d(x,y)
```

```python
ax = plt.axes(projection='3d')
ax.view_init(50, -45)
ax.plot_surface(x, y, z, cmap=cm.coolwarm)
ax.set_xlabel('Eixo x')
ax.set_ylabel('Eixo y')
ax.set_zlabel('Eixo z')
```
